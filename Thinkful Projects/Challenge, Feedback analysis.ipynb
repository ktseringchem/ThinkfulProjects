```python
%matplotlib inline
import numpy as np
import pandas as pd
import scipy
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
review_raw = pd.read_csv('yelp_labelled.txt', delimiter= '\t', header=None)
amazon_raw = pd.read_csv('amazon_cells_labelled.txt', delimiter= '\t', header=None)
review_raw.columns = ['message', 'sentement']
amazon_raw.columns = ['message', 'sentement']
amazon_raw.head()
```

```python
keywords = [ 'only', 'back', 'very', 'over', 'food', 'took', 'horrible', 'don\'t', 'least', 'much', 'excuse', 'down', 'bad', 'any', 'sucks', 'final', 'worst', 'hate', 'hated', 'never','not', 'slow', 'nasty', 'angry', 'didn\'t', 'poor', 'no']
```

```python
for key in keywords:
    # Note that we add spaces around the key so that we're getting the word,
    # not just pattern matching.
    review_raw[str(key)] = review_raw.message.str.contains(
        ' ' + str(key) + ' ',
        case=False
    )
    
review_raw['allcaps'] = review_raw.message.str.isupper()
```

```python
for key in keywords:
    # Note that we add spaces around the key so that we're getting the word,
    # not just pattern matching.
    amazon_raw[str(key)] = amazon_raw.message.str.contains(
        ' ' + str(key) + ' ',
        case=False
    )
    
amazon_raw['allcaps'] = amazon_raw.message.str.isupper()
```

```python
data = review_raw[keywords + ['allcaps'] ]
data2 = amazon_raw[keywords + ['allcaps'] ]
data2.head()
```

```python
target = review_raw['sentement']
target2 = amazon_raw['sentement']
```

```python
# Our data is binary / boolean, so we're importing the Bernoulli classifier.
from sklearn.naive_bayes import BernoulliNB

# Instantiate our model and store it in a new variable.
bnb = BernoulliNB()

# Fit our model to the data.
bnb.fit(data, target)
bnb.fit(data2, target2)

# Classify, storing the result in a new variable.
y_pred = bnb.predict(data)
y_pred2 = bnb.predict(data2)

# Display our results.
print("Number of mislabeled points out of a total {} points : {}".format(
    data.shape[0],
    (target != y_pred).sum()
))
print("Number of mislabeled points in amazon data out of a total {} points : {}".format(
    data2.shape[0],
    (target2 != y_pred2).sum()
))
```
