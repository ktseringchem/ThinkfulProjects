```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
distributionvar = np.random.exponential(1, 100)

plt.axvline(distributionvar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(distributionvar.mean() + distributionvar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(distributionvar.mean() - distributionvar.std(), color='b', linestyle='dashed', linewidth=2)

plt.hist(distributionvar)
plt.show()
```

This is a skewed distribution, therefore does not have true center and variance does not encompase pluse/minus 2 of the SD

```python
distrwald = np.random.wald(1,10000000,100)

plt.axvline(distrwald.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(distrwald.mean() + distrwald.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(distrwald.mean() - distrwald.std(), color='b', linestyle='dashed', linewidth=2)

plt.hist(distrwald)
plt.show()
```

The mean lie in the center when scale goes to infinity, which is inverse of whatever value I pick. Than the graph becomes normaly distributed, but if I bring the mean closer to the scale it becomes a skewed distribution just like the first graph

```python
distrpoisson = np.random.poisson(1000,100)

plt.axvline(distrpoisson.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(distrpoisson.mean() + distrpoisson.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(distrpoisson.mean() - distrpoisson.std(), color='b', linestyle='dashed', linewidth=2)

plt.hist(distrpoisson)
plt.show()
```

There is a true center when the mean is made large but the std is way off the chart. This graph has normal distribution but the variation is too large

```python
distrnorm = np.random.normal(10,1,100)

plt.axvline(distrnorm.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(distrnorm.mean() + distrnorm.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(distrnorm.mean() - distrnorm.std(), color='b', linestyle='dashed', linewidth=2)

plt.hist(distrnorm, bins=20)
plt.show()
```

This is a normal distribution with true center and variance with 65% of the data points within 1 standard deviation.

```python
distrmultinomial = np.random.multinomial(10000,[1/6]*6,1000)

plt.axvline(distrmultinomial.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(distrmultinomial.mean() + distrmultinomial.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(distrmultinomial.mean() - distrmultinomial.std(), color='b', linestyle='dashed', linewidth=2)

plt.hist(distrmultinomial, bins=20)
plt.show()
```

Here I have to control the number of expriment and number of data points in each expriment so naturally as I increase the number of expriments and the number of data point. The distributions begin to look like a normal distribution with center and variance of 65% within the 1 standard deviation of the mean.

```python
distrrayleigh = np.random.rayleigh(1,100)

plt.axvline(distrrayleigh.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(distrrayleigh.mean() + distrrayleigh.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(distrrayleigh.mean() - distrrayleigh.std(), color='b', linestyle='dashed', linewidth=2)

plt.hist(distrrayleigh, bins=20)
plt.show()
```

This hist does not seem to have true cneter and variance does not contain 65% within 1 standard deviation
